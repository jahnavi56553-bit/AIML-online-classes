#
count students in a range and generate marks for 500 students
#
number of students scoring between 60-80  ,  number of students scoring above 90  ,  number of students scoring below 50

```python
import numpy as np

# Generate marks for 500 students (0 to 100)
marks = np.random.randint(0, 101, 500)

# Count students in different ranges
between_60_80 = np.sum((marks >= 60) & (marks <= 80))
above_90 = np.sum(marks > 90)
below_50 = np.sum(marks < 50)

# Display marks (optional)
print("Generated Marks:")
print(marks)

# Display observations
print("\nObservation:")
print("Number of students scoring between 60 and 80:", between_60_80)
print("Number of students scoring above 90:", above_90)
print("Number of students scoring below 50:", below_50)
```

# CSV file

```python
import numpy as np
import pandas as pd
np.random.seed(42)
rows = 1000
df = pd.DataFrame({
"Student_ID": range(1, rows + 1),
"Age": np.random.randint(18, 26, rows),
"Study_Hours": np.random.randint(1, 11, rows),
"Attendance": np.random.randint(60, 101, rows),
"Assignments": np.random.randint(1, 11, rows),
"Previous_Marks": np.random.randint(40, 101, rows),
"Sleep_Hours": np.random.randint(4, 10, rows),
"Internet_Usage": np.random.randint(1, 9, rows),
"Marks": np.random.randint(35, 101, rows)
})
 
print(df.head())
 
df.to_csv("student_performance_1000.csv", index=False)
```

# Probability Operations

```python
# Probabilty of a student scoring more than 80 marks
import pandas as pd
import numpy as np

df = pd.read_csv("student_performance_1000.csv")

prob = len(df[df["Marks"] > 80]) / len(df)
print(prob)

#Probability of attendance above 90%
prob = len(df[df["Attendance"] > 90]) / len(df)
print(prob)

#Probability of study hours greater than 6
prob = len(df[df["Study_Hours"] > 6]) / len(df)
print(prob)

print("Mean:",np.mean(df))
print("Median: ",np.median(df))
```

# Vectorization
Add + 5 marks using numpy for above dataset

```python
import pandas as pd
import numpy as np
df = pd.read_csv("student_performance_1000.csv")

#Addition
marks = df["Marks"].to_numpy()
print("Marks before adding 5: ",marks[:15])
marks = marks+5
print("Adding 5 to marks: ",marks[:15])

#Multiplication
attendance = df["Attendance"].to_numpy()
print("Attendance before multiplying with 2: ",attendance[:10])
attendance =attendance* 2
print("Mutltiplying attendance with 2: ",attendance[:10])

#increase previous marks by 10%
prev_marks = df["Previous_Marks"].to_numpy()
print("Previous marks before adding 10% of marks: ",prev_marks[:10])
prev_marks = prev_marks + prev_marks*0.1
print("Previous marks after adding 10% of marks: ",prev_marks[:10])

#Average of internet usage and sleep hours
sleep_hours = df["Sleep_Hours"].to_numpy()
Internet_usage = df["Internet_Usage"].to_numpy()
Avg = (sleep_hours+Internet_usage)/2
print("Average of sleep hours and internet usage: ",Avg[:10])
```

# 
Create employee database with series and DataFrame

```python
import pandas as pd

# Create Series
emp_id = pd.Series([101, 102, 103, 104])
emp_name = pd.Series(["Jahnavi", "Anuja", "Usha", "Kalyani"])
department = pd.Series(["HR", "IT", "Finance", "Marketing"])
salary = pd.Series([200000, 150000, 100000, 100000])

# Create DataFrame
employee_df = pd.DataFrame({
    "Employee ID": emp_id,
    "Employee Name": emp_name,
    "Department": department,
    "Salary": salary
})

# Display the DataFrame
print("Employee Database")
print(employee_df)
```
